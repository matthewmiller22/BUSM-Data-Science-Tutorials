```python
import os 
import imageio
import scipy.ndimage as ndi
import matplotlib.pyplot as plt
os.listdir('.')
```

```python
import numpy as np
im = imageio.imread(  'ID_0012_AGE_0061_CONTRAST_1_CT.dcm')

plt.imshow(im,cmap = 'gray')
```

```python
sobel_ax0 = ndi.sobel(im, axis=0)
sobel_ax1 = ndi.sobel(im,axis=1)

# Calculate edge magnitude 
edges = edges = np.sqrt(np.square(sobel_ax0) +np.square(sobel_ax1))


# Plot edge magnitude
plt.imshow(edges, cmap='rainbow', vmax=75)
```
